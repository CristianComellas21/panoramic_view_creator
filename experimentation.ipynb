```python
import cv2
import matplotlib.pyplot as plt
```

Load images.

```python


# load images
images = []
for i in range(1, 3):
    img = cv2.imread(f'./images/salon{i}.jpg')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    images.append(img)
```

```python
from algorithm import create_panoramic_view
result = create_panoramic_view(images, match_threshold=0.6)
plt.imshow(result['final_result'])
```

```python
# Pint key points

for i in range(len(result['key_points'])):
    plt.figure()
    cv2.drawKeypoints(images[i], result['key_points'][i], images[i], color=(0, 255, 0), flags=0)
    plt.imshow(images[i])
```

```python
# Plot good matches

plt.imshow(cv2.drawMatchesKnn(images[0], result['key_points'][1], images[1], result['key_points'][0], result['matches'], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS))
```
